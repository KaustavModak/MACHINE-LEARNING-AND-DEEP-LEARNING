## Seperating features and target

```python
X=df.drop(columns='label',axis=1)
y=df['label']
```

```python
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

## Building the Neural Network

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(30,)),
    keras.layers.Dense(20,activation='relu'),
    keras.layers.Dense(2,activation='sigmoid')
])
```

```python
# compiling the NN
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
history=model.fit(X_train_std,y_train,validation_split=0.1,epochs=10)
```

## Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model_accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend(['training_data','validation_data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model_loss')

plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend(['training_data','validation_data'],loc='lower right')
```

## Accuracy of the model on test data

```python
loss,accuracy=model.evaluate(X_test_std,y_test)
print(loss,accuracy)
```

#### in case of y_pred, there are 2 values: loss and accuracy
- loss signifies the probability that this particular datapooint can be 0
- accuracy signifies the probability that this particular datapoint can be 1

model.predict() gives the prediction probabilityof each class for that data point

```python
# argmax function
my_list1=[0.56,0.36]
argmax_val1=np.argmax(my_list1)
print(my_list1)
print(argmax_val1)

my_list2=[0.36,0.56]
armax_val2=np.argmax(my_list2)
print(my_list2)
print(armax_val2)
```

```python
# converting the prediction probability to class labels
y_pred_labels=[np.argmax(i) for i in y_pred]
print(y_pred_labels)
```

## Building the predictive system

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_asarray=np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped=input_data_asarray.reshape(1,-1)

# standardizing the input data
input_data_std=scaler.transform(input_data_reshaped)

prediction=model.predict(input_data_std)
print(prediction)

prediction_label=np.argmax(prediction)

if(prediction_label == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
